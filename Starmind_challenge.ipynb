```python
import pandas as pd
import numpy as np
from Levenshtein import distance as lev

csv_file = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/20210103_hundenamen.csv")
names = csv_file['HUNDENAME'].values

Str1 = "Luca"
distance_list = []
for Str2 in names:
  distance = lev(Str1, Str2) 
  if distance == 1:
    distance_list = np.append(distance_list,Str2)  
print(','.join(distance_list))
```
